```python

from collections import Counter

df['v2_words'] = df['v2'].str.split()

print(df['v2_words'])
all_words = [word for row_words in df['v2_words'] for word in row_words]

# Count the frequency of each word
word_counts = Counter(all_words)

# Get the 5 most common words
top_5_words = word_counts.most_common(5)

print("The 5 most frequent words are:")
for word, count in top_5_words:
    print(f"{word}: {count}")
```

# time complexity 
#####  ,m=tokens , n=stopwords.
##### word_tokenize() -> O(n)
##### filtered_tokens = [token.lower() for token in tokens if token.lower() not in stopwords and token.isalpha()] -> O(m*n) 
##### nltk.FreqDist(filtered_tokens) --> O(m)
##### O(m)+O(m*n)+ O(n) --> O(m*n)

```python
from nltk import word_tokenize
from nltk.stem import WordNetLemmatizer
from nltk.corpus import wordnet
from nltk import pos_tag
from nltk.stem import PorterStemmer


import nltk

# Download required NLTK resources
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('omw-1.4')
nltk.download('averaged_perceptron_tagger')




def Nltk_remove_stopWord(words):
        stopwords = nltk.corpus.stopwords.words('english')
        filtered_tokens = [token.lower() for token in words if token.lower() not in stopwords and token.isalpha()]
        return filtered_tokens
    
def Nltk_Tokenitation(text):
    tokens = nltk.word_tokenize(text)
    return tokens
    
def Nltk_Lemmatization(words):
    lemmatizer = WordNetLemmatizer()
    lemmas = [lemmatizer.lemmatize(word) for word in words]

    return lemmas

def nltk_stemming(words):
    stemmer = PorterStemmer()
    stems = [stemmer.stem(word) for word in words]
    return stems

#nltk_token=Nltk_Tokenitation(all_words)
#print(nltk_token)
#nltk_lemma = Nltk_Lemmatization(nltk_token)
#print(nltk_lemma)
#nltk_stem = nltk_stemming(nltk_token)
#print(nltk_stem)


```

```python
text_expirment="hello, my name is omri chen yosef i am learning new language and i like cats and dogs"
tokeni=Nltk_Tokenitation(text_expirment)
print(tokeni)
removstop=Nltk_remove_stopWord(tokeni)
print(removstop)

Lemmatisation=Nltk_Lemmatization(removstop)

print(Lemmatisation)

print(nltk_stemming(Lemmatisation))
```
